```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Example Data:

```python
data = {
    'Hours_Studied': [1, 2, 3, 4, 5],
    'Exam_Score': [50, 55, 65, 70, 80]
}

df = pd.DataFrame(data)
df
```

Calculate Pearson Correlation:

```python
correlation = df.corr()
print(correlation)
```

Visualize with Heatmap:

```python
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

Spearman Example:

```python
spearman_corr = df.corr(method='spearman')
print(spearman_corr)
```

Univariate Plots (One Variable)
Goal: Explore the distribution of a single variable.

Histogram
Shows the frequency distribution of numerical data.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
tips = sns.load_dataset('tips')
sns.histplot(tips['total_bill'], bins=20, kde=False)
plt.title("Histogram of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

Box Plot
Shows median, quartiles, and outliers.

```python
sns.boxplot(x=tips['total_bill'])
plt.title("Box Plot of Total Bill")
plt.xlabel("Total Bill")
plt.show()
```

KDE Plot (Kernel Density Estimation)
Smooth estimate of the distribution of a variable.

```python
sns.kdeplot(data=tips['total_bill'], shade=True)
plt.title("KDE Plot of Total Bill")
plt.xlabel("Total Bill")
plt.show()
```

Bar Chart (Categorical)
Counts of categories.

```python
sns.countplot(x='day', data=tips)
plt.title("Bar Chart of Days")
plt.xlabel("Day")
plt.ylabel("Count")
plt.show()
```

Bivariate Plots (Two Variables)
Goal: Explore relationships between two variables.

Scatter Plot
Best for continuous vs. continuous data.

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title("Scatter Plot of Tip vs Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

Line Plot
Ideal for time series or continuous trends.

```python
sns.lineplot(x='total_bill', y='tip', data=tips)
plt.title("Line Plot of Tip vs Total Bill")
plt.show()
```

Hexbin Plot (with matplotlib)
Used for large datasets to show density.

```python
plt.hexbin(tips['total_bill'], tips['tip'], gridsize=25, cmap='Blues')
plt.colorbar(label='Count')
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Hexbin Plot")
plt.show()
```

Grouped Bar Plot
Comparing categories across a group.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset('tips')

# Create grouped bar plot with 'smoker' as the hue
sns.barplot(x='day', y='total_bill', hue='smoker', data=tips)

# Add title
plt.title("Grouped Bar Plot by Smoker and Day")

# Show the plot
plt.show()
```

 Multivariate Plots (More than Two Variables)
Goal: Visualize complex relationships.

Pair Plot
All possible scatter plots among numerical variables.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
tips = sns.load_dataset('tips')

# Create pair plot with 'smoker' as the hue
sns.pairplot(tips, hue='smoker')

# Add title to the plot
plt.suptitle("Pair Plot with Hue: Smoker vs Non-Smoker", y=1.02)

# Show the plot
plt.show()
```

Heatmap (Correlation Matrix)
Shows pairwise correlation with colors.

```python
corr = tips.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Heatmap of Correlations")
plt.show()

```

3D Scatter Plot (with plotly)

```python
import plotly.express as px

fig = px.scatter_3d(tips, x='total_bill', y='tip', z='size',
                    color='sex', title='3D Scatter Plot')
fig.show()
```

Bubble Chart (size as third variable)

```python
import plotly.express as px

# Load dataset
tips = px.data.tips()

# Create bubble chart with 'smoker' as the color
fig = px.scatter(tips, x='total_bill', y='tip',
                 size='size', color='smoker', hover_data=['day'],
                 title="Bubble Chart: Tip vs Total Bill (Smoker vs Non-Smoker)")

# Show the plot
fig.show()
```

Violin Plot with Hue
Combines boxplot and KDE for distribution per category.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset('tips')

# Create violin plot with 'smoker' as the hue
sns.violinplot(x='day', y='total_bill', hue='smoker', data=tips, split=True)

# Add title
plt.title("Violin Plot of Total Bill by Day and Smoker Status")

# Show the plot
plt.show()
```
